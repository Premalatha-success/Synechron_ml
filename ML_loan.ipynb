```python
median_loanamount=df["LoanAmount"].median()
median_loanamount
```

```python
df["LoanAmount"]=df["LoanAmount"].replace(np.nan,median_loanamount)
```

```python
df.isnull().sum()
```

```python
### replace for loan amount term and credit history
```

```python
median_term=df["Loan_Amount_Term"].median()
```

```python
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].replace(np.nan,median_term)
```

```python
median_credit=df["Credit_History"].median()
```

```python
df["Credit_History"]=df["Credit_History"].replace(np.nan,median_credit)
```

```python
df.isnull().sum()
```

```python
mode_gender=df["Gender"].mode()[0]
mode_gender
```

```python
df["Gender"]=df["Gender"].replace(np.nan,mode_gender)
```

```python
mode_married=df["Married"].mode()[0]
```

```python
mode_employed=df["Self_Employed"].mode()[0]
```

```python
mode_dep=df["Dependents"].mode()[0]
```

```python
df["Married"]=df["Married"].replace(np.nan,mode_married)
```

```python
df["Dependents"]=df["Dependents"].replace(np.nan,mode_dep)
```

```python
df["Self_Employed"]=df["Self_Employed"].replace(np.nan,mode_employed)
```

```python
df.isnull().sum()
```

```python
df.sample(10)
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

```python
df.info()
```

```python
Y=df["Loan_Status_Y"]
X=df.drop("Loan_Status_Y",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train)
print(model_lr.score(X_train,Y_train))
print(model_lr.score(X_test,Y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_kn=KNeighborsClassifier()
model_kn.fit(X_train,Y_train)
print(model_kn.score(X_train,Y_train))
print(model_kn.score(X_test,Y_test))
```

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
print(model_svc.score(X_train,Y_train))
print(model_svc.score(X_test,Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
model_dt.fit(X_train,Y_train)
print(model_dt.score(X_train,Y_train))
print(model_dt.score(X_test,Y_test))
```

```python
from sklearn.ensemble import BaggingClassifier
model_bg=BaggingClassifier()
model_bg.fit(X_train,Y_train)
print(model_bg.score(X_train,Y_train))
print(model_bg.score(X_test,Y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
model_ab=AdaBoostClassifier()
model_ab.fit(X_train,Y_train)
print(model_ab.score(X_train,Y_train))
print(model_ab.score(X_test,Y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_gb=GradientBoostingClassifier()
model_gb.fit(X_train,Y_train)
print(model_gb.score(X_train,Y_train))
print(model_gb.score(X_test,Y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier()
model_rf.fit(X_train,Y_train)
print(model_rf.score(X_train,Y_train))
print(model_rf.score(X_test,Y_test))
```
