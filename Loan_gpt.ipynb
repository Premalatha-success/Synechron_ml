```python
import matplotlib.pyplot as plt
import seaborn as sns

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Exclude 'Loan_ID' from categorical_cols if it exists
if 'Loan_ID' in categorical_cols:
    categorical_cols.remove('Loan_ID')

print("\n--- Univariate Analysis for Numerical Columns (Histograms and Boxplots) ---")
for col in numerical_cols:
    print(f"\nAnalyzing numerical column: {col}")
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col].dropna())
    plt.title(f'Histogram of {col}')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[col].dropna())
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()

print("\n--- Univariate Analysis for Categorical Columns (Countplots) ---")
for col in categorical_cols:
    print(f"\nAnalyzing categorical column: {col}")
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Define the target variable
target_variable = 'Loan_Status'

# Exclude 'Loan_ID' from categorical columns if it exists, as it's an identifier
if 'Loan_ID' in categorical_cols:
    categorical_cols.remove('Loan_ID')

# Remove the target variable from the list of features to plot against itself
if target_variable in categorical_cols:
    categorical_cols.remove(target_variable)

print("\n--- Bivariate Analysis: Loan_Status vs. Categorical Columns (Countplots) ---")
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, hue=target_variable, palette='coolwarm')
    plt.title(f'Loan_Status Distribution by {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

print("\n--- Bivariate Analysis: Loan_Status vs. Numerical Columns (Boxplots) ---")
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x=target_variable, y=col, palette='viridis')
    plt.title(f'{col} Distribution by Loan_Status')
    plt.tight_layout()
    plt.show()
```

```python
### Data Preparation-> Drop insignificant column(Loan_ID),Treat missing values,Encoding
```

```python
df=df.drop("Loan_ID",axis=1)
```

```python
print("Missing values before imputation:")
print(df.isnull().sum())

# Impute categorical columns with mode
for col in ['Gender', 'Married', 'Dependents', 'Self_Employed']:
    if col in df.columns:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Impute numerical columns with median
for col in ['LoanAmount', 'Loan_Amount_Term','Credit_History']:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)


```

```python
print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```python
print("\nDataFrame info after imputation:")
df.info()
```

```python
# Identify categorical columns to encode (excluding the target variable 'Loan_Status')
categorical_features = df.select_dtypes(include='object').columns.tolist()
if 'Loan_Status' in categorical_features:
    categorical_features.remove('Loan_Status')

# Apply one-hot encoding to categorical features
df = pd.get_dummies(df, columns=categorical_features, drop_first=True)

# Encode the target variable 'Loan_Status' (Y/N to 1/0)
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})
```

```python

print("DataFrame after encoding categorical columns:")
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Separate features (X) and target (y)
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=1)


# Initialize and train the Logistic Regression model
model_lr = LogisticRegression(solver="liblinear")
model_lr.fit(X_train, y_train)
print(model_lr.score(X_train,y_train))
print(model_lr.score(X_test,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_kn=KNeighborsClassifier(n_neighbors=21,weights="uniform")
model_kn.fit(X_train,y_train)
print(model_kn.score(X_train,y_train))
print(model_kn.score(X_test,y_test))
```

```python
from sklearn.svm import SVC #kernel,C,gamma
model_svm=SVC()
model_svm.fit(X_train,y_train)
print(model_svm.score(X_train,y_train))
print(model_svm.score(X_test,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier(max_depth=5)
model_dt.fit(X_train,y_train)
print(model_dt.score(X_train,y_train))
print(model_dt.score(X_test,y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
model_ab=AdaBoostClassifier(learning_rate=0.05)
model_ab.fit(X_train,y_train)
print(model_ab.score(X_train,y_train))
print(model_ab.score(X_test,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_gb=GradientBoostingClassifier(learning_rate=0.01)
model_gb.fit(X_train,y_train)
print(model_gb.score(X_train,y_train))
print(model_gb.score(X_test,y_test))
```

```python
from sklearn.ensemble import BaggingClassifier
model_bgf=BaggingClassifier(max_samples=0.05)
model_bgf.fit(X_train,y_train)
print(model_bgf.score(X_train,y_train))
print(model_bgf.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier(max_depth=3)
model_rf.fit(X_train,y_train)
print(model_rf.score(X_train,y_train))
print(model_rf.score(X_test,y_test))
```
